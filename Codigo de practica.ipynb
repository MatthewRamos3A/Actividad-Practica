# =========================================
# SISTEMA SIMPLE DE GESTIÓN DE TAREAS
# =========================================
# Este programa permite:
# 1. Agregar tareas
# 2. Ver tareas
# 3. Eliminar tareas
# 4. Salir del programa
# =========================================

```python
# Lista para almacenar las tareas
tareas = []
```

```python
def agregar_tarea():
    tarea = input("Ingrese una nueva tarea: ")
    tareas.append(tarea)
    print("Tarea agregada correctamente.")
```

```python
def mostrar_tareas():

    if len(tareas) == 0:
        print("No hay tareas registradas.")

    else:
        print("\nLista de tareas:")

        for i, tarea in enumerate(tareas, start=1):
            print(f"{i}. {tarea}")
```

```python
def eliminar_tarea():

    mostrar_tareas()

    if len(tareas) > 0:

        numero = int(input("Ingrese el número de la tarea a eliminar: "))

        if 1 <= numero <= len(tareas):
            tarea_eliminada = tareas.pop(numero - 1)
            print(f"Tarea eliminada: {tarea_eliminada}")

        else:
            print("Número inválido.")
```

```python
while True:

    print("\n====== MENÚ ======")
    print("1. Agregar tarea")
    print("2. Mostrar tareas")
    print("3. Eliminar tarea")
    print("4. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_tarea()

    elif opcion == "2":
        mostrar_tareas()

    elif opcion == "3":
        eliminar_tarea()

    elif opcion == "4":
        print("Programa finalizado.")
        break

    else:
        print("Opción inválida.")
```
